```python
df_benin['Date'] = df_benin['Timestamp'].dt.date
daily_ghi_benin = df_benin.groupby('Date')['GHI'].sum()

plt.figure(figsize=(15,5))
daily_ghi_benin.plot()
plt.title("Daily Total GHI  Benin")
plt.xlabel("Date")
plt.ylabel("GHI (Wh/m)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_benin['Hour'] = df_benin['Timestamp'].dt.hour
hourly_ghi_benin = df_benin.groupby('Hour')['GHI'].mean()

plt.figure(figsize=(10,4))
hourly_ghi_benin.plot(kind='bar', color='green')
plt.title("Average GHI per Hour  Benin")
plt.xlabel("Hour of Day")
plt.ylabel("Avg GHI")
plt.grid(True)
plt.tight_layout()
plt.show()
```
