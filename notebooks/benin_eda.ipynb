```python
# Check missing values by column
df_benin.isnull().sum()
```

```python
# Drop rows where Timestamp is missing (if any)
df_benin = df_benin.dropna(subset=['Timestamp'])

# Fill numeric columns with their mean
numeric_cols = df_benin.select_dtypes(include=['float64', 'int64']).columns

df_benin[numeric_cols] = df_benin[numeric_cols].fillna(df_benin[numeric_cols].mean())

# Check again to confirm
print(df_benin.isnull().sum())
```

```python
# Check number of duplicates before removal
duplicates_before = df_benin.duplicated().sum()
print(f"Duplicate rows before removal: {duplicates_before}")

# Flag duplicates
df_benin['IsDuplicate'] = df_benin.duplicated()

# Count flagged duplicates
print(f"Flagged duplicates: {df_benin['IsDuplicate'].sum()}")
```

```python
from scipy.stats import zscore

# Columns to check for outliers
columns_to_check = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'WS', 'WSgust']

# Compute Z-scores for selected columns
z_scores = df_benin[columns_to_check].apply(zscore)

# Flag rows with any Z-score > 3
df_benin['Outlier'] = (z_scores.abs() > 3).any(axis=1)

# Count how many rows were flagged
print(f"Number of outliers flagged: {df_benin['Outlier'].sum()}")
```

```python
# Columns you care about
key_columns = ['GHI', 'DNI', 'DHI', 'ModA', 'ModB', 'Tamb', 'RH', 'WS', 'WSgust', 'WSstdev', 'WD', 'WDstdev', 'BP', 'TModA', 'TModB']

# Impute missing values with median
for col in key_columns:
    if df_benin[col].isna().sum() > 0:
        median_value = df_benin[col].median()
        df_benin[col].fillna(median_value, inplace=True)
        print(f"Filled missing values in {col} with median: {median_value}")
    else:
        print(f"No missing values in {col}")
```

```python
# Export cleaned DataFrame to CSV
df_benin.to_csv('../data/benin_clean.csv', index=False)
print(" Cleaned Benin data exported to '../data/benin_clean.csv'")
```

```python
import matplotlib.pyplot as plt

# Calculate means before removing outliers
moda_mean_before = df_benin.loc[~df_benin['Outlier'], 'ModA'].mean()
modb_mean_before = df_benin.loc[~df_benin['Outlier'], 'ModB'].mean()

# Drop outliers for cleaning impact comparison
df_cleaned = df_benin[~df_benin['Outlier']].copy()

# Calculate means after removing outliers
moda_mean_after = df_cleaned['ModA'].mean()
modb_mean_after = df_cleaned['ModB'].mean()

# Prepare data for plotting
impact_df = pd.DataFrame({
    'ModA': [moda_mean_before, moda_mean_after],
    'ModB': [modb_mean_before, modb_mean_after]
}, index=['Before Cleaning', 'After Cleaning'])

# Plot
impact_df.plot(kind='bar', figsize=(8, 5), color=['skyblue', 'orange'])
plt.title('Average ModA and ModB Before vs After Cleaning')
plt.ylabel('Average Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Scatter Plots for Correlation & Relationship Analysis
# This section visualizes the relationships between key variables to identify trends or correlations.

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(16, 10))

# GHI vs DNI
plt.subplot(2, 2, 1)
sns.scatterplot(data=df_cleaned, x='GHI', y='DNI', alpha=0.4, color='orange')
plt.title("GHI vs DNI")
plt.xlabel("Global Horizontal Irradiance (GHI)")
plt.ylabel("Direct Normal Irradiance (DNI)")

# Tamb vs GHI
plt.subplot(2, 2, 2)
sns.scatterplot(data=df_cleaned, x='Tamb', y='GHI', alpha=0.4, color='green')
plt.title("Tamb vs GHI")
plt.xlabel("Ambient Temperature (Tamb C)")
plt.ylabel("GHI")

# Tamb vs RH
plt.subplot(2, 2, 3)
sns.scatterplot(data=df_cleaned, x='Tamb', y='RH', alpha=0.4, color='blue')
plt.title("Tamb vs RH")
plt.xlabel("Ambient Temperature (Tamb C)")
plt.ylabel("Relative Humidity (RH %)")

# BP vs Tamb
plt.subplot(2, 2, 4)
sns.scatterplot(data=df_cleaned, x='BP', y='Tamb', alpha=0.4, color='purple')
plt.title("BP vs Tamb")
plt.xlabel("Barometric Pressure (BP)")
plt.ylabel("Ambient Temperature (Tamb C)")

plt.tight_layout()
plt.show()
```

```python
# Convert WD to direction bins (every 30)
df_cleaned['WD_bin'] = pd.cut(df_cleaned['WD'], bins=np.arange(0, 361, 30), right=False)

# Compute mean wind speed per bin
wind_rose_data = df_cleaned.groupby('WD_bin', observed=True)['WS'].mean().reset_index()
wind_rose_data['Direction'] = wind_rose_data['WD_bin'].apply(lambda x: float(x.left + 15))
wind_rose_data = wind_rose_data.dropna()
wind_rose_data['Direction'] = pd.to_numeric(wind_rose_data['Direction'])

# Convert direction to radians
theta = np.deg2rad(wind_rose_data['Direction'])
radii = wind_rose_data['WS']

# --- Wind Rose Diagram ---
plt.figure(figsize=(8, 8))
ax = plt.subplot(111, polar=True)
ax.set_theta_zero_location("N")
ax.set_theta_direction(-1)

bars = ax.bar(theta, radii, width=np.deg2rad(30), color='skyblue', alpha=0.7)
plt.title("Wind Rose Diagram (Benin)")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample a smaller set if the dataset is too large (optional)
sample_df = df_cleaned.sample(n=1000, random_state=42)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    sample_df['Tamb'], 
    sample_df['GHI'], 
    s=sample_df['RH'],  # Bubble size by RH
    alpha=0.5, 
    c=sample_df['RH'], 
    cmap='viridis',
    edgecolors='w'
)

plt.xlabel("Ambient Temperature (Tamb)")
plt.ylabel("Global Horizontal Irradiance (GHI)")
plt.title("Bubble Chart: Tamb vs GHI (Bubble Size = RH)")
plt.colorbar(label='Relative Humidity (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Bubble Chart: Tamb vs GHI

This chart visualizes the relationship between ambient temperature (`Tamb`) and global horizontal irradiance (`GHI`), with bubble size representing relative humidity (`RH`). The chart helps to identify how GHI varies with temperature under different humidity levels.

- **X-axis**: Ambient Temperature (Tamb)
- **Y-axis**: Global Horizontal Irradiance (GHI)
- **Bubble Size & Color**: Relative Humidity (RH)
